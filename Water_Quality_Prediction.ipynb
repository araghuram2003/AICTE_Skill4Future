## Listing Column Names

```python
df.columns
```

```python
pollutants=['O2', 'NO3', 'NO2', 'SO4','PO4', 'CL']
```

```python
pollutants
```

### Now we are drop the missing values using - dropna()

```python
df = df.dropna(subset=pollutants)
```

```python
df.head()
```

## Feature and Target Section 
##### x is the feature, it consist of id and year, which act as a independent varables
##### y is a traget, it consist of list of pollutants varable, which act as a dependent varaibles to be predicted 

```python
x = df[['id','year']]
y= df[pollutants]
```

## Encoding 

We are apply **one-hot encoding** to the `'id'` column, which represents 22 station IDs.  
By setting `drop_first=True`, one of the encoded columns is dropped to avoid multicollinearity (dummy variable trap).  
This results in 21 binary columns representing station IDs, making the data suitable for machine learning models


```python
x_encoded= pd.get_dummies(x,columns=['id'], drop_first=True)
```

### Now we are splitting the data into training and testing sets, with 80% for training and 20% for the testing

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)
```

### Training a multi-output random forest model on the training data


```python
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(x_train, y_train)
```

### Making predictions on the test data


```python
y_pred= model.predict(x_test)
```

### Model Performance Evaluation 

This code evaluates model performance on test data by calculating and printing the Mean Squared Error (MSE) and R-squared (R2) score for each pollutant.


```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

## Here we are Making a Prediction

- Gave a station ID and year as input.
- Created input data and applied one-hot encoding.
- Matched it with the training data format.
- Used the trained model to predict pollutant levels.
- Printed the predicted values for each pollutant.


```python
station_id = '10'
year_input = 2025

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(x_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[x_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```

## Now Saving the Model

- Used `joblib` to save the trained model to a file (`pollution_model.pkl`).
- Also saved the list of feature column names (`model_columns.pkl`).
- This lets us reuse the model later without retraining.

```python
import joblib

joblib.dump(model, 'pollution_model.pkl')
joblib.dump(x_encoded.columns.tolist(), "model_columns.pkl")
print('Model and cols structure are saved')
```
